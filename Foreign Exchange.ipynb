## Exploratory Data Analysis(EDA)

```python
# Droping column

fxr = fxr.drop(['Unnamed: 0'], axis=1)

fxr
```

**Succesfully Dropped the column**

```python
#Coverting dtype from object to datetime

fxr['Time Serie'] = pd.to_datetime(fxr['Time Serie'], format = '%Y-%m-%d' )
fxr.info()
```

**Successfully changed dtype from object to datetime**

```python
#Coverting dtype from object to float and filling the object type as null

cols = fxr.columns.drop('Time Serie')
fxr[cols] = fxr[cols].apply(pd.to_numeric, errors='coerce')
fxr.info()
```

**Successfully changed dtype from object to float and filling the object type as null**

```python
#Checking null values

fxr.isnull().sum()
```

**We can drop the null values because it is few in numbers**

```python
#Droping the null values with row reference

fxr.dropna(inplace=True)
fxr
```

**Now the data looks neat**

```python
# Checking null values

fxr.isnull().sum()
```

**No null values**

```python
# Statistical View 

fxr.describe()
```

**We are ready to go for Visualization**

```python
# We need to convert dataset values from different currencies to USD
# We will take inverse of the currency so that we will get the currencies in USD

fxr = pd.concat([fxr["Time Serie"], (1/fxr.drop(["Time Serie"],axis=1))], axis=1, join='inner')
fxr
```

**We have changed all the currencies to US$**

```python
# Plotting Currently HIGHEST Currency

plt.figure(figsize=(20,7))
sns.lineplot( x="Time Serie", 
              y=(fxr.iloc[-1].drop('Time Serie').sort_values(ascending=False).head(1).index[0]),
              data=fxr )
plt.xlabel("Time Series")
plt.title("HIGHEST CURRENCY")
plt.show()
```

**UK Pounds are in highest currently**

```python
# Plotting Currently LOWEST Currency

plt.figure(figsize=(20,7))
sns.lineplot( x="Time Serie", 
              y=(fxr.iloc[-1].drop('Time Serie').sort_values().head(1).index[0]),
              data=fxr )
plt.xlabel("Time Series")
plt.title("LOWEST CURRENCY")
plt.show()
```

**Korean Won are in lowest currently**

```python
# Checking relationship between lowest and highest trending currencies

plt.figure(figsize=(20,7))
sns.regplot(x = 'UNITED KINGDOM - UNITED KINGDOM POUND/US$',
            y = 'KOREA - WON/US$',
            data = fxr.sample(100,random_state=2))
plt.show()
```

**We can see the connection between the currencies**

```python
# CURRENCY IMPROVEMENT IN DOLLARS

improve = pd.DataFrame({"CURRENCY IMPROVEMENT IN DOLLARS":((fxr.iloc[-1].drop('Time Serie')-fxr.iloc[0].drop('Time Serie'))
                                                 .sort_values(ascending=False))})
plt.figure(figsize=(20,10))
sns.barplot(x='CURRENCY IMPROVEMENT IN DOLLARS', y=improve.index,data=improve,palette="coolwarm")
plt.show()
```

**We can see postive growth currencies in Blue, negative growth currencies in Red**

```python
# Currently trending Currencies according to last date of 2019

current = pd.DataFrame({"Dollars":fxr.iloc[-1].drop('Time Serie').sort_values(ascending=False)})

plt.figure(figsize=(20,10))
sns.barplot(x="Dollars", y=current.index,data=current,palette="Dark2")
plt.show()
```

**We can see from powerfull to weakest currencies**

```python
# Correlation Heatmap Plot between currencies

plt.figure(figsize=(15,6))
sns.heatmap(fxr.corr(),cmap="tab10",annot=True)
plt.show()
```

**We can observe Euro and Denmark have perfect positive correlation Why? Check in the below link** https://en.wikipedia.org/wiki/Denmark_and_the_euro

```python
# Checking with Scatter plot

plt.figure(figsize=(15,6))
sns.scatterplot(x=fxr["DENMARK - DANISH KRONE/US$"],y=fxr["EURO AREA - EURO/US$"])
plt.show()
```

**We can see a Pure Simple Regression line**

```python
# Some of the Currencies having high correlation

plt.figure(figsize=(15,6))
sns.pairplot(fxr[['AUSTRALIA - AUSTRALIAN DOLLAR/US$','EURO AREA - EURO/US$',
                  'NEW ZEALAND - NEW ZELAND DOLLAR/US$','CANADA - CANADIAN DOLLAR/US$',
                  'SINGAPORE - SINGAPORE DOLLAR/US$','DENMARK - DANISH KRONE/US$',
                  'SWEDEN - KRONA/US$','NORWAY - NORWEGIAN KRONE/US$']])
plt.show()
```

**They have some similarity as we can observe**

```python
#MAX,MIN and last day price of the Currencies with there Years

pd.DataFrame({"MAXIMUM PRICE":fxr.groupby(pd.DatetimeIndex(fxr['Time Serie']).year)[cols].mean().max(),
              "YEAR OF MAX":fxr.groupby(pd.DatetimeIndex(fxr['Time Serie']).year)[cols].mean().idxmax(),
              "MINIMUM PRICE":fxr.groupby(pd.DatetimeIndex(fxr['Time Serie']).year)[cols].mean().min(),
              "YEAR OF MIN":fxr.groupby(pd.DatetimeIndex(fxr['Time Serie']).year)[cols].mean().idxmin(),
              "LAST DAY PRICE":fxr.drop('Time Serie',axis =1).iloc[-1]})
```

**Checking MAXIMUM and MINIMUM for the particular currencies with year they have appeared and also last day price of the currencies**
